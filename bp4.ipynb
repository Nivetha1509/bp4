```python
#1.Write a Python Program to Find the Factorial of a Number?
a=int(input())
prod=1
for i in range (1,a+1):
    prod=prod*i
print(prod)
```

```python
#2. Write a Python Program to Display the multiplication Table?
a=int(input())
for i in range (1,11):
    mul=i*a
    print("{}*{}={}".format(i,a,mul))
```

```python
#3. Write a Python Program to Print the Fibonacci sequence?
length=int(input("How many terms?"))
a=0
b=1
print(b)
for i in range(1,length):
    new=a+b
    a=b
    b=new
    print(new)
    
    
```

```python
#4. Write a Python Program to Check Armstrong Number?
n=int(input())
orginal_number=n
while(orginal_number>0):
    rem=orginal_number%10
    res=rem*rem*rem
    orginal_number//=10
if(res==orginal_number):
    print("{} is a Armstrong Number".format(n))
else:
    print("{} is not a Armstrong Number".format(n))
    
```

```python
#5. Write a Python Program to Find Armstrong Number in an Interval?
start=int(input())
end=int(input())
for i in range (start,end+1):
    ord=len(str(i))
    orginal_number=i
    res=0
    while(orginal_number>0):
        rem=orginal_number%10
        res+=rem**ord
        orginal_number//=10
    if(res==i):
        print("{} is a Armstrong Number".format(i))
    else:
        print("{} is not a Armstrong Number".format(i))
```

```python
#6. Write a Python Program to Find the Sum of Natural Numbers?
num=int(input())
sum=0
if(num<=0):
    print("Enter the correct Number")
else:
    while(num>0):
        sum=sum+num
        num=num-1
print(sum)        
```
